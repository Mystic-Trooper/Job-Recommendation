```python
import pandas as pd
import json
import random

job_description_dataset = json.load(
    open("Dataset\dataset_job_description.json","r"))
resume_dataset = pd.read_json("Dataset\dataset_resume.json")


```

```python
print(len(job_description_dataset))
relavent_skills=[]
for resumeData in resume_dataset.values:  # first two resumes
    relavent_skills.extend(resumeData[0]['careerjunction_za_skills'])
# print(relavent_skills)  # skills

```

```python
for i in range(0, len(job_description_dataset)):
# for i in range(0,1):
    random_count = random.randint(8, 20)
    random_n_skills = random.sample(relavent_skills, random_count)
    str="";
    for i in range(0, len(random_n_skills)):
        str+=f"{random_n_skills[i]}"
        if(i != len(random_n_skills)-1):
            str+=" , "
    job_description_dataset[i]["skills"] = str
    print(i)
```

```python
json.dump(job_description_dataset, open(
    "Dataset\dataset_job_description.json", "w"))
```
